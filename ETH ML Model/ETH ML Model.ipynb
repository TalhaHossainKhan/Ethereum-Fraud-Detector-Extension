```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler

# Load and preprocess data
df = pd.read_csv("3.csv")
positive_sample = df[df["Flag"] == 1]
negative_sample = df[df["Flag"] == 0].sample(len(positive_sample), random_state=23)
data = pd.concat([negative_sample, positive_sample], axis=0)
y = data["Flag"]
X = data.iloc[:,2:]

# Fill NaN values with mean
for col in X.columns:
    X[col] = X[col].fillna(X[col].mean())

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the Random Forest model
rf_model = RandomForestClassifier(n_estimators=150, criterion="entropy", bootstrap=True, random_state=23)
rf_model.fit(X_train_scaled, y_train)
```
